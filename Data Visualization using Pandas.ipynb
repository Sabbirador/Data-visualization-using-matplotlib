```python
plt.figure(figsize=(10,5))

plt.title('Gas Price Over Time in USD',fontdict={'fontweight':'bold','fontsize':18,'color':'g'})

plt.plot(gas.Year,gas['USA'],'b.-',label='USA')
plt.plot(gas.Year,gas['Canada'],'g.-',label='Canada')
plt.plot(gas.Year,gas['South Korea'],'r.-',label='S korea')
plt.plot(gas.Year,gas['Australia'],'y.-',label='Australia')

#plt.xticks([1990,1993,1996,1999,2002,2005,2008])
plt.xticks(gas.Year[: :2].tolist())



plt.xlabel('Year')
plt.ylabel('US dollar')


plt.legend()

plt.show()


```
